```python
#Using Bayes, what are the best two predictors that someone died EXCLUDING GENDER?
#use regression and then use bayes or do various probabilities with bayes

import pandas as pd
import numpy as ny

Location = "titanic3_step1.csv"
df = pd.read_csv(Location)
```

```python
#incorporate embarked column
def embarked_to_numeric(x):
    if x=='Q':
        return 0
    if x=='S':
        return 1
    if x=='C':
        return 2
    
df['embarked'] = df['embarked'].apply(embarked_to_numeric)
```

```python
df.corr()
```

```python
#regression before bayes not taking into account gender
#class, fare and accomodation are not independent of one another
#can only use 1 in the niave bayes, so only test 1 in the regression

#accomodation = 0.283
#pclass = -0.26
#hasfamily = 0.236
#fare = 0.202
#embarked = 0.1729
#parch = 0.131
#age = -0.10
#sibsp = 0.0209

#based on correlations hasfamily and either accomodation, pclass, or fare look like will work the best for regression
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ hasfamily + accomodation -1', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ hasfamily + pclass -1', data=df).fit()
result.summary()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ hasfamily + fare -1', data=df).fit()
result.summary()
```

```python
#trying out next runner up embarked
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ embarked + accomodation -1', data=df).fit()
result.summary()
```

```python
#trying out next runner up embarked
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ embarked + pclass -1', data=df).fit()
result.summary()
#not good pclass p value too high
```

```python
#trying out next runner up embarked
import statsmodels.formula.api as sm
result = sm.ols(formula='survived ~ embarked + fare -1', data=df).fit()
result.summary()
```

```python
#based on the regression exluding gender the top 3 regressions are:
#1. embarked and accomodation               R-squared = 0.491
#2. embarked and fare                       R-squared = 0.474
#3. hasfamily and accomodation              R-squared = 0.460

#not exactly sure how to split up the fare so with try #1 and #3
```

```python
#family
hasfamily_Died = len(df[(df['survived']==0) & (df['hasfamily']==1)])

#accomodation
cabin_Died = len(df[(df['survived']==0) & (df['accomodation']==1)])
steerage_Died = len(df[(df['survived']==0) & (df['accomodation']==0)])

#embarked
embarkedQ_Died = len(df[(df['survived']==0) & (df['embarked']==0)])
embarkedS_Died = len(df[(df['survived']==0) & (df['embarked']==1)])
embarkedC_Died = len(df[(df['survived']==0) & (df['embarked']==2)])

total_Died = len(df[df['survived']==0])
```

```python
#probability that died using two best predictors
def prob_d_twoBest(x,y):
    return x*y
```

```python
def prob_a_b(a,b):
    return a/b
```

```python
#embarkedQ and cabin accomodation died
print(prob_d_twoBest(prob_a_b(embarkedQ_Died,total_Died),prob_a_b(cabin_Died,total_Died)))

#1.2% change of dying if embarkedQ and were in cabin
```

```python
#embarkedQ and steerage accomodation died
print(prob_d_twoBest(prob_a_b(embarkedQ_Died,total_Died),prob_a_b(steerage_Died,total_Died)))

#4.4% change of dying if embarkedQ and were in steerage
```

```python
#embarkedS and cabin accomodation died
print(prob_d_twoBest(prob_a_b(embarkedS_Died,total_Died),prob_a_b(cabin_Died,total_Died)))

#17% change of dying if embarkedS and were in cabin
```

```python
#embarkedS and cabin steerage died
print(prob_d_twoBest(prob_a_b(embarkedS_Died,total_Died),prob_a_b(steerage_Died,total_Died)))

#61% change of dying if embarkedC and were in steerage
```

```python
#embarkedC and cabin accomodation died
print(prob_d_twoBest(prob_a_b(embarkedC_Died,total_Died),prob_a_b(cabin_Died,total_Died)))

#3.4% chance dying if embarkedC and were in cabin
```

```python
#embarkedC and cabin steerage died
print(prob_d_twoBest(prob_a_b(embarkedC_Died,total_Died),prob_a_b(steerage_Died,total_Died)))

#12% change of dying if embarkedC and were in steerage
```

```python
#hasfamily and cabin accomodation died
print(prob_d_twoBest(prob_a_b(hasfamily_Died,total_Died),prob_a_b(cabin_Died,total_Died)))

#8.9% chance of dying if had family and were in cabin
```

```python
#hasfamily and steerage accomodation died
print(prob_d_twoBest(prob_a_b(hasfamily_Died,total_Died),prob_a_b(steerage_Died,total_Died)))

#31% chance of dying if had family and were in steerage
```

```python
#highest probabilities of dying are:
#1. 61% for embarkedS and steerage
#2. 31% for hasfamily and steerage

#both have steerage in common so being a lower class passenger who stayed in steerage is a major factor contributing to if died
```
