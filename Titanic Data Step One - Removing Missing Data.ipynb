```python
import pandas as pd
import numpy  as np

Location = "titanic3.csv"

#saved the excel file from Canvas as a csv file
#csv file in same folder as Titanic Data Step One

df = pd.read_csv(Location)
```

```python
#hasfamily: 0 or 1
df.loc[df["sibsp"] + df["parch"] == 0, "hasfamily"] = 0
df.loc[df["sibsp"] + df["parch"] >= 1, "hasfamily"] = 1

df.to_csv("titanic3_step1.csv", index=False, header=True)
```

```python
#status: 0 lost at sea or 1 body recovered or 2 survived

#lost at sea means no values in both boat and body columns for a person
df.loc[df["boat"].isnull() & df["body"].isnull(), "status"] = 0

#a body was recovered if there is a numerical entry in the body column
df.loc[df["body"].notnull(), "status"] = 1

#could have had tickets but didn't get on?
#survival means on boat 1-20 or A,B,C,D
#boat data type is string due to letters
#some passengers have multiple boat entries
#it is because it is not known for sure which boat they were on or they may have been on more than one boat
#instead of looking for numbers and letters, just check if emtpy or not
df.loc[df["boat"].notnull(), "status"] = 2

df.to_csv("titanic3_step1.csv", index=False, header=True)
```

```python
#accomodation: 0 for steerage or 1 for cabin
df.loc[df["cabin"].isnull(), "accomodation"] = 0
df.loc[df["cabin"].notnull(), "accomodation"] = 1

df.to_csv("titanic3_step1.csv", index=False, header=True)
```

```python
#Then remove the rows that are truly missing data.

#removing rows with missing age data
df = df.drop(df[df["age"].isnull()].index)

#removing rows with missing fare data
#passengers with fare == 0 had complimentary passage
df = df.drop(df[df["fare"].isnull()].index)

#rows with no data in the cabin column mean the passenger was in steerage; don't remove

#removing rows with missing embarked data
df = df.drop(df[df["embarked"].isnull()].index)

#rows with no data in the boat column and no data in the body column were lost at sea; don't remove

#removing rows with missing home.dest data
df = df.drop(df[df["home.dest"].isnull()].index)

#other columns have all data
#hasfamily, status, and accomodations created by me so non-emtpy

df.to_csv("titanic3_step1.csv", index=False, header=True)
```
