```python
#Take the titanic dataset that you cleaned up and look at it for correlations so that you can do a regression
#that allows you to predict whether someone would survive based on their attributes.

import pandas as pd
import numpy as ny

Location = "titanic3_step1.csv"
df = pd.read_csv(Location)
```

```python
#incorporate sex column
def sex_to_numeric(x):
    if x=='female':
        return 1
    else:
        return 0
    
df['sex'] = df['sex'].apply(sex_to_numeric)
```

```python
#incorporate embarked column
def embarked_to_numeric(x):
    if x=='Q':
        return 0
    if x=='S':
        return 1
    if x=='C':
        return 2
    
df['embarked'] = df['embarked'].apply(embarked_to_numeric)
```

```python
#survival based on status
def stat_surv(x):
    if x== 2:
        return 1
    if x!= 2:
        return 0
    
df['stat_surv'] = df['status'].apply(stat_surv)
```

```python
#looking to see what factors affect survival based on status
#survived column is different than status column per examination during step 1
#survived is more correlated to survival based on status than status
#compare survived with survival based on status

#try sex, accomodation, and passenger class
#maybe hasfamily and fare
#maybe combinations of parent and child for hasfamily
#not sure about embarked

# pclass + sex + age + sibsp + parch + fare + embarked + accomodation
# pclass + sex + age + hasfamily + fare + embarked + accomodation

#tried various combinations and below is what looks best
```

```python
import statsmodels.formula.api as sm
result = sm.ols(formula='stat_surv ~ sex + accomodation + age + embarked -1', data=df).fit()
result.summary()
#remove fare and parch
#removed sibsp
#tried hasfamily and removed
```
